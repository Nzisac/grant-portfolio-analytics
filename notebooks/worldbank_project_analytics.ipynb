Create dataframe from raw data

```python
import pandas as pd

worldbank_df = pd.read_excel("../data/raw/world_bank_projects_africa.xlsx")
```

```python
worldbank_df.head()
```

```python
worldbank_df.info()
```

```python
worldbank_df.columns = (worldbank_df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
worldbank_df.info()

```

```python
worldbank_df.dropna(subset=["implementing_agency", "borrower", "theme_1", "theme_2"], inplace=True)
worldbank_df.info()
worldbank_df["country"].value_counts()
```

Save the cleaned file to Processed


```python
target_countries = ["Republic of Kenya", "United Republic of Tanzania", "Republic of Uganda", "Republic of Rwanda"]
worldbank_filtered_df = worldbank_df[worldbank_df["country"].isin(target_countries)]
worldbank_filtered_df.info()
```

```python
print(worldbank_filtered_df.columns.tolist())
print(worldbank_filtered_df.shape)   
```

```python
worldbank_filtered_df.to_csv("../data/processed/world_bank_projects_africa_cleaned.csv", index=False)
```
